**Convert to Dataframe**

---

<br><br>

**(1) iterate**

```python
# 상단의 with open과 연결; data에 할당
dd = []
for i in data['features']:
    dd.append(i['candidate'])

pd.DataFrame(dd).head()
```



---


<br><br>

**(2) read_json**

```python
df = pd.read_json('/employee_data.json')

# all features의 value를 list >> column명 'candidate'
bn = pd.DataFrame(df.features.values.tolist())['candidate']

# 확인용01
df.features.values

# 확인용02
df.features.values.tolist()

# 새로운 DataFrame 만들기01; 이건 효과가 없음(전과 동일한 결과물)
pd.DataFrame(bn)

# 새로운 DataFrame 만들기02; ✭from_records 넣어주기✭ << 이게 적용
pd.DataFrame.from_records(bn)

pd.DataFrame.from_records(bn).head()
```

**(3) json_normalize**

```python
# (2)와 동일
df = pd.read_json('/employee_data.json')
bn = pd.DataFrame(df.features.values.tolist())['candidate']

# normalize
pd.json_normalize(bn).head()
```
